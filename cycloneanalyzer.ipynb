```python
from PIL import Image, ImageDraw
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import math
import numpy as np
im1 = Image.open(r'D:\5th sem\EE604A\EE604-Assign1-pictures\1.jpg')
im2 = Image.open(r'D:\5th sem\EE604A\EE604-Assign1-pictures\2.jpg')
im3 = Image.open(r'D:\5th sem\EE604A\EE604-Assign1-pictures\3.jpg')
draw = ImageDraw.Draw(im2)

plt.subplot(3, 1, 1)
plt.imshow(im1)
plt.axis("off")
plt.title("im1")

plt.subplot(3, 1,2)
plt.imshow(im2)
plt.axis("off")
plt.title("im2")

plt.subplot(3, 1, 3)
plt.imshow(im3)
plt.axis("off")
plt.title("im3")
#Show all imported Images
plt.show()
imageSizeW, imageSizeH = im1.size
#Print Image Parameters we are working with
print(imageSizeW)
print(imageSizeH)
```

# Getting center points

```python
#We need Red pixel from this Image to find loaction of center, rather than doing it manually we will do looping
#It is for core of both cyclones and a for Ambali and b for Belna
RedPixels = []
for l in range(0,3):
    for j in range(1, imageSizeH):
        for i in range(1, imageSizeW):
            if l==0:
                R,G,B= im1.getpixel((i, j))
                if R>240 and B<12 and G<12:
                    RedPixels.append([i, j])
            if l==1:
                R,G,B= im2.getpixel((i, j))
                if R>240 and B<12 and G<12:
                    RedPixels.append([i, j])
            if l==2:
                R,G,B= im3.getpixel((i, j))
                if R>240 and B<12 and G<12:
                    RedPixels.append([i, j])
print(len(RedPixels))
Red = []
t=0
for i in range(0,6):
    n=0
    x=0
    y=0
    x1=RedPixels[t][0]
    y1=RedPixels[t][1]
    for k in range(t,len(RedPixels)):
        if abs(x1-RedPixels[t][0]) > 10 and abs(y1-RedPixels[t][1]) > 10:
            break
        n=n+1
        x=x+RedPixels[t][0]
        y=y+RedPixels[t][1]
        t=t+1
    print(n)
    x=int(x/n)
    y=int(y/n)
    Red.append([x,y])
print(Red)
```

```python
#Feeding them in more memoriable variable for ease of use
b1=Red[0]
a1=Red[1]
b2=Red[2]
a2=Red[3]
b3=Red[4]
a3=Red[5]
```

```python
BluePixels = []
#It is for Two location in blue
for j in range(1, imageSizeH):
    for i in range(1, imageSizeW):
        R,G,B= im1.getpixel((i, j))
        if B>240 and R<12 and G<12:
            BluePixels.append([i, j])
print(len(BluePixels))
Blue = []
t=0
for i in range(0,2):
    n=0
    x=0
    y=0
    x1=BluePixels[t][0]
    y1=BluePixels[t][1]
    for k in range(t,len(BluePixels)):
        if abs(x1-BluePixels[t][0]) > 10 and abs(y1-BluePixels[t][1]) > 10:
            break
        n=n+1
        x=x+BluePixels[t][0]
        y=y+BluePixels[t][1]
        t=t+1
    print(n)
    x=int(x/n)
    y=int(y/n)
    Blue.append([x,y])
print(Blue)
```

```python
#Feeding them in more memoriable variable for ease of use
Mauritius=Blue[0]
Reunion=Blue[1]
```

```python
GreenPixels = []
#It is 3 location in Madagascar
for j in range(1, imageSizeH):
    for i in range(1, imageSizeW):
        R,G,B= im1.getpixel((i, j))
        if G>240 and B<12 and R<12:
            GreenPixels.append([i, j])
print(len(GreenPixels))
Green = []
t=0
for i in range(0,3):
    n=0
    x=0
    y=0
    x1=GreenPixels[t][0]
    y1=GreenPixels[t][1]
    for k in range(t,len(GreenPixels)):
        if abs(x1-GreenPixels[t][0]) > 10 and abs(y1-GreenPixels[t][1]) > 10:
            break
        n=n+1
        x=x+GreenPixels[t][0]
        y=y+GreenPixels[t][1]
        t=t+1
    print(n)
    x=int(x/n)
    y=int(y/n)
    Green.append([x,y])
print(Green)
```

```python
#Feeding them in more memoriable variable for ease of use
Antsiranana=Green[0]
Antananarivo=Green[1]
Benonoka=Green[2]
```

# Finding distance

```python
Mad=math.dist(Antsiranana,Benonoka)
print(Mad)
#Mad is distance in Pixel between north Antsiranana (green) to south Benonoka (green) which is 1500 kilometers
```

```python
Uni=math.dist(Mauritius,Reunion)
print(Uni)
#Uni is distance in Pixel between Reunion and Mauritius (blue) and are separated by 250 kilometres
```

```python
Bel1=math.dist(b1,b2)
print(Bel1)
Bel2=math.dist(b2,b3)
print(Bel2)
#Bel1 and Bel2 are distance between cyclones in intermediate image of Belna Cyclone in (Image1,Image2) and (Image3,Image2) respectively
```

```python
Amb1=math.dist(a1,a2)
print(Amb1)
Amb2=math.dist(a2,a3)
print(Amb2)
#Amb1 and Amb2 are distance between cyclones in intermediate image of Ambali Cyclone in (Image1,Image2) and (Image3,Image2) respectively
```

# Scaling fator

```python
sf1= 1500/Mad
print(sf1)
#Scaling factor for distance for Antsiranana and Benonoka
```

```python
sf2= 250/Uni
print(sf2)
#Scaling factor for distance for Reunion and Mauritius
```

```python
#Average of Scaling factor for non linear scaling
sf = (sf1+sf2)/2
print(sf)
```

# Calculation for points

```python
# Trajectory of cyclone "Belna" can be represented by straight line (y-y1)/(x-x1) = (y3-y1)/(x3-x1) i.e. 12x+7y−3523=0
# Trajectory of cyclone "Ambali" can be represented by straight line (y-y2)/(x-x2) = (y4-y2)/(x4-x2) i.e. 7x+11y−5841=0
# Solving the above two simulataneous equations with two unknowns
A = np.array([[12, 7], [7, 11]])
B = np.array([3523, 5841])
Apojuncture = np.linalg.solve(A,B)
```

```python
Apojuncture
```

```python
distapojtocap = (math.dist(Apojuncture,Antananarivo))*sf
print(distapojtocap)
#Distance between Apojuncture and Antananarivo is printed below
```

```python
time = 55/60 # time to capture img2 and img3
speed_Belna = (math.dist(b2, b3)*sf)/time # Speed of Belna
speed_Ambali = (math.dist(a2, a3)*sf)/time # Speed of Ambali
```
